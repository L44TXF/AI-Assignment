### 1.2: Describe the models used 

Logistic regression

Logistic regression is a statistical method for analysing a dataset in which there are one or more independent variables that determine an outcome. The outcome is measured with a dichotomous variable (in which there are only two possible outcomes). This model is used to predict a binary outcome yes/no given a set of independent variables. Relating to our problem where our mushroom can only be poisenous or not. This model predicts the probability of the occurance of poisenous by fitting data to a function. Predictions are mapped to be between 0 and 1 through the logistic function, which means that predictions can be interpreted as class probabilities.The models themselves are still "linear," so they work well when your classes are linearly separable (i.e. they can be separated by a single decision surface). 

The strengths of this model are as follows:
- Outputs have a nice probabilistic interpretation, and the algorithm can be regularized to avoid overfitting.
- Logistic models can be updated easily with new data using stochastic gradient descent.

The weaknesses are as follows: 
- Logistic regression tends to underperform when there are multiple or non-linear decision boundaries.
- They are not flexible enough to naturally capture more complex relationships.



Decision tree  

Decision trees are a type of supervised machine learning where the data is continuously split according to a certain parameter. The tree can be explained by two entities, namely deccision nodes and leaves. The leaves are the decisions or the final outcomes. And the decision nodes are where the data is split. There are two types of decision trees: classification and regression. In my work I have used the classification trees as the outcome was a binary variable which was either poisenous or edible.

The strengths of this model are as follows: 
- As with regression, classification tree ensembles also perform very well in practice.
- They are robust to outliers, scalable, and able to naturally model non-linear decision boundaries thanks to their hierarchical structure.

The weaknesses are as follows: 
- Unconstrained, individual trees are prone to overfitting, but this can be alleviated by ensemble methods.

**What encoding method would you use (does it matter?)**

I was taught two encoding methods: label encoding and one hot encoding. For my dataset I used one hot encoding as a breifly explained above. Learners like logistic regression, distance based methods such as kNN, support vector machines, tree based methods etc. in sklearn needs numeric arrays. Features having string values cannot be handled by these learners. This is why the strings and other values are changed to ints.

Categorical data are variables that contain label values rather than numberic valies. The dataset used in my assignment is an example of this where all the data is nominal. 

Label incoding is a feature of Sklearn. It works by assigning each unique category value an integer value. For example, "blue" is 1, "green" is 2 and "red" is 3. Noticably, the integer values have a natural ordered relationshop between each other and machine learning algorithms are may be able to understand and harness this relationship. An example where this would be useful is assigning place values in a race where there is "first" (1), "second" (2) and "third"(3) where the integer has weighted value. 

However, for categorical variables where this ordinal relationship exists the integer encoding is not enough. In fact, using label encoding and allowing the model to assume a natural ordering between categories may result in poor performance or unexpected results. Ideally one hot encoding should be used. This is where the integer encoded variable is removed and a new binary variable is added for each unique integer value. This does resolve the weighting and performance issue but brings in issues itself. For example, the data does not scale well. For nominal data with various varibles new columns would need to be created and each row would require a binary value to be added to it. On a large scale this is computationally straining and may cause worse performance issues.

```
#Found a good way to display the results as a class
def print_score(classifier,X_train,y_train,X_test,y_test,train=True):
    if train == True:
        print("Training results:\n")
        print('Accuracy Score: {0:.4f}\n'.format(accuracy_score(y_train,classifier.predict(X_train))))
        print('Classification Report:\n{}\n'.format(classification_report(y_train,classifier.predict(X_train))))
        print('Confusion Matrix:\n{}\n'.format(confusion_matrix(y_train,classifier.predict(X_train))))
        res = cross_val_score(classifier, X_train, y_train, cv=10, n_jobs=-1, scoring='accuracy')
        print('Average Accuracy:\t{0:.4f}\n'.format(res.mean()))
        print('Standard Deviation:\t{0:.4f}'.format(res.std()))
    elif train == False:
        print("Test results:\n")
        print('Accuracy Score: {0:.4f}\n'.format(accuracy_score(y_test,classifier.predict(X_test))))
        print('Classification Report:\n{}\n'.format(classification_report(y_test,classifier.predict(X_test))))
        print('Confusion Matrix:\n{}\n'.format(confusion_matrix(y_test,classifier.predict(X_test))))
```
